```python
import json, os

Task_file = "Tasks.json"
completed_task_file = "completed_task.json"

if not os.path.exists(Task_file):
    with open(Task_file, "w") as f:
        json.dump([], f)

def load_task():
    with open(Task_file, "r") as f:
        return json.load(f)

def save_task(tasks):
    with open(Task_file, "w") as f:
        json.dump(tasks, f)

def add_tasks():
    task_name = input("Enter the task name:")
    tasks = load_task()
    tasks.append({"task": task_name , "status": "pending"})
    save_task(tasks)
    print(f"Task {task_name} added successfully!")

def view_tasks():
    tasks = load_task()
    if tasks:
        print("To do list")
        for index, task in enumerate(tasks, start=1):
            print(f"{index}. {task['task']} - {task['status']}")
    else:
        print("No tasks found.")
def update_task():
    tasks = load_task()
    view_tasks()
    try:
        task_index = int(input("Enter the task number to update:"))-1
        if 0 <= task_index < len(tasks):
            new_status = input("Enter the new status(completed/pending):")
            tasks[task_index]["status"] = new_status
            save_task(tasks)
            print(f"Task '{tasks}'updated successfully!")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Invalid input. Please enter a valid task number.")

def implement_date():
    tasks = load_task()
    view_tasks()
    try:
        task_index = int(input("Enter the task number to update:"))-1
        if 0 <= task_index < len(tasks):
            new_date = input("Enter the new date:")
            tasks[task_index]["date"] = new_date
            save_task(tasks)
            print(f"Task '{tasks}'updated successfully!")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Invalid input. Please enter a valid task number.")

def delete_task():
    tasks = load_task()
    view_tasks()
    try:
        task_index = int(input("Enter the task number to delete:"))-1
        if 0 <= task_index < len(tasks):
            deleted_tasks= tasks.pop(task_index)
            save_task(tasks)
            print(f"Task '{deleted_tasks['task']}'deleted successfully!")
        else:
            print("Invalid task number.")

    except:
        print("Invalid input. Please enter a valid task number.")

def completed_task():
    tasks = load_task()
    view_tasks()

    try:
        completed_tasks = [task for task in tasks if task["status"] == "completed"]
        with open(completed_task_file, "w") as f:
                json.dump(completed_tasks,f)
        print(f"Task '{len(completed_tasks)}'is completed successfully!")

    except ValueError:
        print("Invalid input. Please enter a valid task number.")
    except Exception as e:
        print(f"An error occurred: {e}")


def display_menu():
    print("\nTo-Do List Menu:")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Update Task")
    print("4. Task implement date")
    print("5. Delete Task")
    print("6. Completed Task")
    print("7. Exit")


while True:
    display_menu()
    choice = input("Enter your choice (1/2/3/4/5/6/7):")
    if choice == "1":
        add_tasks()
    elif choice == "2":
        view_tasks()
    elif choice == "3":
        update_task()
    elif choice == "4":
        implement_date()
    elif choice == "5":
        delete_task()
    elif choice == "6":
        completed_task()
    elif choice == "7":
        print("Exiting the program.")
        break
    else:
        print("Invalid choice. Please select a valid option (1/2/3/4/5/6/7).")


```
